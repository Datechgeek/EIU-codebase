```python
#Importing Relevant Libraries
import pandas as pd 
import matplotlib.pyplot as plt 
import plotly.express as px 
```

```python


df = pd.read_csv(r"C:\Users\USER\OneDrive\VS Code Repos\EIU codebase\EIU-codebase\Data\tomato_retail_sales_clean.csv", parse_dates=['Date'])
df.info()
```

```python
# Seasonal Impact Analysis for "Total Sales Value (NGN)" 
seasonal_impact =df.groupby("Season")["Total Sales Value (NGN)"].mean().sort_values(ascending=False)
seasonal_impact
```

```python
#Visualizing Seasonal Impact Analysis for "Total Sales Value (NGN)"
seasonal_impact.plot(kind="bar", xlabel="Season", ylabel="Total Sales Value (NGN)", title="Distribution of Total Sales per season" );
```

```python
# Seasonal Impact Analysis for "Quantity Sold (kg)" 
seasonal_impact_Quantity =df.groupby("Season")["Quantity Sold (kg)"].mean().sort_values(ascending=False)
seasonal_impact_Quantity
```

```python
#Visualizing Seasonal Impact Analysis for  "Quantity Sold (kg)" 
seasonal_impact_Quantity.plot(kind="bar", xlabel="Season", ylabel="Quantity Sold (kg)", title="Distribution of Quantity Sold per season " );
```

```python

# Seasonal Impact Analysis for "Unit Price (NGN)" 
seasonal_impact_Unit_Price =df.groupby("Season")["Unit Price (NGN)"].mean().sort_values(ascending=False)
seasonal_impact_Unit_Price
```

```python
#Visualizing Seasonal Impact Analysis for "Unit Price (NGN)"
seasonal_impact_Unit_Price.plot(kind="bar", xlabel="Season", ylabel="Unit Price (NGN)", title="Distribution of Unit Price (NGN) per season" );
```

```python
#Boxplot for visualizing Unit Price in Dry Season
df_dry_season = df[df["Season"] == "Dry Season"]
plt.boxplot(df_dry_season["Unit Price (NGN)"], vert=False)
plt.xlabel("Unit Price (NGN)"),
plt.title("Distribution of Unit Price in Dry Season");
```

```python
#Boxplot for visualizing Quantity Sold in Dry Season
df_dry_season = df[df["Season"] == "Dry Season"]
plt.boxplot(df_dry_season["Quantity Sold (kg)"], vert=False)
plt.xlabel("Quantity Sold (kg)"),
plt.title("Distribution of Quantity Sold in Dry Season");
```

```python
#Boxplot for Visualizing Quantity sold in Rainy Season
df_rainy_season = df[df["Season"] == "Rainy Season"]
plt.boxplot(df_rainy_season["Quantity Sold (kg)"], vert=False)
plt.xlabel("Quantity Sold (kg)"),
plt.title("Distribution of Quantity Sold in Rainy Season");
```

```python
#Boxplot for Visualizing Unit Price in Rainy Season
df_rainy_season = df[df["Season"] == "Rainy Season"]
plt.boxplot(df_rainy_season["Unit Price (NGN)"], vert=False)
plt.xlabel("Unit Price (NGN)"),
plt.title("Distribution of Unit Price in Rainy Season");
```

CONCLUSION: For "Total Sales Value (NGN), the Dry Season saw more  revenue in Sales compared to the Rainy Season
For "Quantity Sold (kg)" , the Dry Season saw more Sales compared to the Rainy Season
For "Unit Price (NGN)", the Rainy saw an increase in unit price compared to the dry season.
