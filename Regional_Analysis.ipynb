```python
#Importing Relevant Libraries
import pandas as pd 
import matplotlib.pyplot as plt 
import plotly.express as px 
```

```python
df = pd.read_csv(r"C:\Users\USER\OneDrive\VS Code Repos\EIU codebase\EIU-codebase\Data\tomato_retail_sales_clean.csv", parse_dates=['Date'])
df.info()
```

```python
#Regional Analysis for Price 
regional_store_price = df.groupby("Store Location")["Unit Price (NGN)"].mean().sort_values(ascending=False)
regional_store_price
```

```python
#Visualizing Regional Analysis for "Unit Price (NGN)"
regional_store_price.plot(
    kind="bar", xlabel="Store Location", ylabel="Unit Price (NGN)",
    title="Difference in Unit Price Per Region"
    );
```

```python
#Regional Analysis for Total Sales 
regional_store_sales = df.groupby("Store Location")["Total Sales Value (NGN)"].mean().sort_values(ascending=False)
regional_store_sales
```

```python
#Visualizing Regional Analysis for "Total Sales Value (NGN)"
regional_store_sales.plot(
    kind="bar", xlabel="Store Location", ylabel="Total Sales Value (NGN)",
    title="Difference in Total Sales Value (NGN)"
    );
```

CONCLUSION: There is no much difference in Demand or Price of Goods Across various Regions. This has been discovered by the various aggregations and further confirmed by the visualizations.
