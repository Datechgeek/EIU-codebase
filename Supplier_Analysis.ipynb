```python
#Importing Relevant Libraries
import pandas as pd 
import matplotlib.pyplot as plt 
import plotly.express as px 
```

```python
#Load Data Into DataFrame
df = pd.read_csv(r"C:\Users\USER\OneDrive\VS Code Repos\EIU codebase\EIU-codebase\Data\tomato_retail_sales_clean.csv", parse_dates=['Date'])
df.info()
```

```python
#Inspecting Supplier name column
df["Supplier Name"].value_counts()
```

```python
#Understanding Correlation Between Supplier Cost and Transport cost
supplier_transport_correlation = df["Supplier Cost (NGN)"].corr(df["Transport Cost (NGN)"])
print("Correlation Coefficient: ",supplier_transport_correlation)
```

```python
#Visualizing Correlation Between Supplier Cost and Transport cost using Scatter Plot
plt.scatter(x=df["Supplier Cost (NGN)"], y=df["Transport Cost (NGN)"])
plt.xlabel("Supplier Cost (NGN)")
plt.ylabel("Transport Cost (NGN)")
plt.title("Correlation Between Supplier Cost and Transport cost");
```

CONCLUSION: There is a very weak correlation between supplier Cost and Transport cost as shown in the visualization above.

```python
#Understanding Correlation Between  Transport cost and Unit Price (NGN)
transport_price_correlation = df["Transport Cost (NGN)"].corr(df["Unit Price (NGN)"])
print("Correlation Coefficient: ",transport_price_correlation)

```

```python
#Visualizing Correlation Between  Transport cost and Unit Price (NGN)
plt.scatter(x=df["Transport Cost (NGN)"], y=df["Unit Price (NGN)"])
plt.xlabel("Transport Cost (NGN)")
plt.ylabel("Unit Price (NGN)")
plt.title("Correlation Between Transport cost and Unit Price");
```

CONCLUSION: There is a very weak correlation between Transport cost and Unit Price (NGN) as shown in the visualization above.
