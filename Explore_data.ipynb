CONDUCTING EDA ON THE CLEANED DATAFRAME
The next step after Cleaning our data is to explore and better understand the various features and their relationships so we can unlock insights useful for the next step in our project lifecycle.

NUMERICAL DATA: we are going to explore key numerical features in our DataFrame.

```python
#Getting the descriptive statistics of "Unit Price (NGN)".
df["Unit Price (NGN)"].describe()
```

```python
# Visualizing distribution using a histogram
plt.hist(df["Unit Price (NGN)"]),
plt.xlabel("Unit Price (NGN)"),
plt.ylabel("Frequency"),
plt.title("Distribution of Unit Price");
```

```python
# Visualizing distribution using a Boxplot
plt.boxplot(df["Unit Price (NGN)"], vert=False)
plt.xlabel("Unit Price (NGN)"),
plt.title("Distribution of Unit Price");
```

CONCLUSION: There is a normal distribution (Bell Curve) for the "Unit Price (NGN)", this was shown in the descriptive statistics by the minimal difference between the mean and the median(50th Percentile) and further confirmed by the visualization.
