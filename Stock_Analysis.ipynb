```python
#Importing Relevant Libraries
import pandas as pd 
import matplotlib.pyplot as plt 
import plotly.express as px
```

```python
#Load Data Into DataFrame
df = pd.read_csv(r"C:\Users\USER\OneDrive\VS Code Repos\EIU codebase\EIU-codebase\Data\tomato_retail_sales_clean.csv", parse_dates=['Date'])
df.info()
```

```python
#Opening Stock In Dry Season
df_dry_season = df[df["Season"] == "Dry Season"]
print("Opening Stock in dry season:", df_dry_season["Opening Stock (kg)"].mean())
```

```python
#Closing Stock In Dry Season
df_dry_season = df[df["Season"] == "Dry Season"]
print("Closing Stock in dry season:", df_dry_season["Closing Stock (kg)"].mean())
```

```python
#Opening Stock In Rainy Season
df_dry_season = df[df["Season"] == "Rainy Season"]
print("Opening Stock in Rainy season:", df_dry_season["Opening Stock (kg)"].mean())
```

```python
#Closing Stock In Rainy Season
df_dry_season = df[df["Season"] == "Rainy Season"]
print("Closing Stock in Rainy season:", df_dry_season["Closing Stock (kg)"].mean())
```

```python
# Calculate daily stock change
df["Stock Change (kg)"] = df["Closing Stock (kg)"] - df["Opening Stock (kg)"]

# Plot stock levels over time

plt.plot(df["Date"], df["Opening Stock (kg)"], label="Opening Stock (kg)")
plt.plot(df["Date"], df["Closing Stock (kg)"], label="Closing Stock (kg)")
plt.xlabel("Date")
plt.ylabel("Stock (kg)")
plt.title("Stock Levels Over Time")
plt.legend()
```

```python
# Analyze trends with respect to seasonality
df.groupby("Season")["Stock Change (kg)"].mean().plot(kind="bar")
plt.xlabel("Season")
plt.ylabel("Average Stock Change (kg)")
plt.title("Average Stock Change by Season");
```

```python
# Analyze trends with respect to pricing changes
plt.scatter(df["Unit Price (NGN)"], df["Stock Change (kg)"])
plt.xlabel("Unit Price (NGN)")
plt.ylabel("Stock Change (kg)")
plt.title("Stock Change vs. Unit Price")
plt.show()
```

```python
#Correlation between Stock Change and Price
stock_change_price = df["Unit Price (NGN)"].corr(df["Stock Change (kg)"])
print("Correlation Coefficient: ",stock_change_price)
```

CONCLUSION:
    1. There is no Noticeable change in Stock Change Over Time
    2. There is  a very Noticeable Difference in Stock Change with respect to season, With the Dry Season seeing more Stocks over the rainy season.
    3. There is a very weak correlation between stock change and Unit Price
